**What is DBSCAN?**

-Unsupervised clustering algorithm that groups data based on density.

-Detects arbitrary-shaped clusters and outliers (noise).

-Does not require the number of clusters (K) to be specified.

**üß† Key Concepts**
Term	Description
-Core Point	Has at least min_samples points within distance eps
-Border Point	Not a core, but within eps of a core point
-Noise Point	Not a core or border ‚Üí considered outlier (labeled -1)

**üîß Parameters**
Parameter	Meaning
-eps	Radius of neighborhood (distance threshold)
-min_samples	Minimum number of points to form a dense region

‚öôÔ∏è How DBSCAN Works
Start with all points unvisited.

**For each point**:

-Find neighbors within eps

-If ‚â• min_samples, it becomes a core point

-Form a cluster and expand by connecting other core points

**Advantages**

-Mark unclustered points as noise.
üö´ No need to pre-specify number of clusters (unlike K-Means)

‚úÖ Detects noise/outliers

‚úÖ Works with arbitrarily-shaped clusters

‚úÖ Deterministic (no randomness)


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import sklearn.datasets
```

```python
iris = sklearn.datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
X=df
scaler=StandardScaler()
X=scaler.fit_transform(X)
X_df=pd.DataFrame(X,columns=iris.feature_names)
X_df.head()

```

```python
db=DBSCAN(eps=0.5,min_samples=5)
```

```python
predictions=db.fit_predict(X)
```
